```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

lab2_data = sns.load_dataset('titanic')
lab2_data
```

```python
sns.barplot(data=lab2_data, x='class', y='survived')
```

```python
sns.barplot(data=lab2_data, x='class', y='survived', hue='who')
```

```python
import pandas as pd
lab2_data.loc[(lab2_data['who'] == 'child') & (lab2_data['sex'] == 'male'), 'who'] = 'boy'
lab2_data.loc[(lab2_data['who'] == 'child') & (lab2_data['sex'] == 'female'), 'who'] = 'girl'

sns.barplot(data=lab2_data, x='class', y='survived', hue='who')
```

```python
sns.barplot(data=lab2_data, x='class', y='survived', hue='who')
```

```python
sns.barplot(data=lab2_data, x='class', y='survived', hue='who', capsize=0.05, errorbar=('ci', 90))
```

```python
sns.barplot(data=lab2_data, x='survived', y='age', hue='who')
```

```python
lab2_data.loc[(lab2_data['survived'] == 0), 'survived_str'] = 'no'
lab2_data.loc[(lab2_data['survived'] == 1),  'survived_str'] = 'yes'
lab2_data
```

```python
sns.barplot(data=lab2_data, x='survived_str', y='age', hue='who')

import matplotlib.pyplot as plt
plt.xlabel("survived")
```

```python
sns.barplot(data=lab2_data, x='survived_str', y='age', hue='who', estimator=np.mean, capsize=0.05)

plt.xlabel("survived")
```

```python
sns.boxplot(data=lab2_data, x='survived', y='age', hue='who')
```

```python
sns.stripplot(data=lab2_data, x='survived', y='age', hue='who', jitter=0.3, alpha=0.5, dodge='true')
```
