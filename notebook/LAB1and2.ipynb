## Get data from drive using wgen and manipulate it using wc and grep

```python
%%bash
wget -nv -O IGHJ_Mouse.fa 'https://drive.google.com/uc?export=download&id=19OjK8m9M9WwUZpB-5FLSu7RWeQhRbYKO'
wget -nv -O IGHJ_Human.fa 'https://drive.google.com/uc?export=download&id=1kUzbiYfBOYQDXUaUzJ4NEGJN9ysPulBz'
wget -nv -O IGHJ_Zebrafish.fa 'https://drive.google.com/uc?export=download&id=1m0VB08CBfl6GZ9rfCtmDyh6PI7elHcak'
```

```python
%%bash
cat > headerExtractor
grep -h ">" *fa > headers.txt # the hedaers from files with extentions *fa are written in headers.txt
```

```python
%%bash

chmod u+x headerExtractor
```

```python
%%bash

./headerExtractor
```

```python
%%bash
cat > seqMeasurer

for fasta in *fa; do
  grep -h ">" "$fasta" >> lengths.txt # the headers are appended
  grep -v ">" "$fasta" > temp # not the headears are collected in temp variable ...
  wc -m temp >> lengths.txt # and its number of chars are appended to lengthts
done
```

```python
%%bash

chmod u+x seqMeasurer
```

```python
%%bash

./seqMeasurer
```
